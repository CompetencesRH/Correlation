```python
import pandas as pd

# Exemple de DataFrame (valeurs numériques)
df = pd.DataFrame({
    "satisfaction": [4.2, 3.8, 4.5, 3.1, 4.9],
    "performance":  [3.9, 3.5, 4.0, 2.8, 4.6],
    "anciennete":   [2.5, 3.1, 7.0, 0.8, 12.2],
    "age":          [28, 32, 41, 25, 50]
})

# Calculer la matrice de corrélation complète
matrice_corr = df.corr(method="pearson")
print(matrice_corr)


##df.corr() calcule automatiquement toutes les corrélations deux à deux entre colonnes numériques.

##La diagonale vaut toujours 1.0 (chaque variable est parfaitement corrélée à elle‑même).

# Corrélation de Pearson (par défaut)
matrice_pearson = df.corr(method="pearson")

# Corrélation de Spearman (rang, moins sensible aux valeurs extrêmes)
matrice_spearman = df.corr(method="spearman")

# Corrélation de Kendall
matrice_kendall = df.corr(method="kendall")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(matrice_corr, annot=True, cmap="coolwarm", center=0)
plt.title("Matrice de corrélation complète")
plt.show()
##annot=True = affiche le chiffre dans chaque case,

##cmap="coolwarm" = couleur rouge pour corrélation positive.
```
